```python
import pandas as pd
import numpy as np
import sklearn
import warnings
warnings.simplefilter("ignore")
```

```python
df = pd.read_csv('train.csv')
df.head(10)
```

```python
X_test = pd.read_csv('test.csv')
X_test
```

```python
df.describe()
```

```python
df = df.drop('Name', axis=1)
```

```python
y_train = df['Survived']
X_train = df.drop('Survived', axis=1)
```

```python
X_train.drop(['PassengerId', 'Cabin', 'Ticket'], axis=1, inplace=True)
```

```python
X_test.drop(['Name', 'PassengerId', 'Cabin', 'Ticket'], axis=1, inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cat_cols = X_train.columns[X_train.dtypes == object].tolist()
for i in cat_cols:
    le.fit(X_train[i].fillna('N'))
    X_train[i] = le.transform(X_train[i].fillna('N'))
    X_test[i] = le.transform(X_test[i].fillna('N'))
X_train
```

```python
X_train['Age'].fillna(X_train['Age'].median(), inplace=True)
X_test['Age'].fillna(X_test['Age'].median(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].median(), inplace=True)
```

```python
X_train['Age_bin'] = pd.cut(X_train['Age'], bins=[0, 12, 20, 40, 100], labels=['A', 'B', 'C', 'D'])
X_train['Fare_bin'] = pd.cut(X_train['Fare'], bins=[0, 7.91, 14.45, 31, 120], labels=['A', 'B', 'C', 'D'])
```

```python
X_test['Age_bin'] = pd.cut(X_test['Age'], bins=[0, 12, 20, 40, 100], labels=['A', 'B', 'C', 'D'])
X_test['Fare_bin'] = pd.cut(X_test['Fare'], bins=[0, 7.91, 14.45, 31, 120], labels=['A', 'B', 'C', 'D'])
```

```python
X_train.drop(['Age', 'Fare'], axis=1, inplace=True)
X_test.drop(['Age', 'Fare'], axis=1, inplace=True)
X_train
```

```python
cat_cols = ['Age_bin', 'Fare_bin']
df = pd.get_dummies(X_train[cat_cols])
df.drop(['Age_bin_A', 'Fare_bin_A'], axis=1, inplace=True)
X_train.drop(['Age_bin', 'Fare_bin'], axis=1, inplace=True)
X_train = X_train.join(df)
X_train
```

```python
df = pd.get_dummies(X_test[cat_cols])
df.drop(['Age_bin_A', 'Fare_bin_A'], axis=1, inplace=True)
X_test.drop(['Age_bin', 'Fare_bin'], axis=1, inplace=True)
X_test = X_test.join(df)
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score

DC = DecisionTreeClassifier()
params_1 = {'max_depth': range(1, 12)}
grid_1 = GridSearchCV(DC, params_1, cv=5, scoring='accuracy')
grid_1.fit(X_train, y_train)

print("Best depth = %.4f" % grid_1.best_params_["max_depth"])
print("Best Accuracy = %.4f" % grid_1.best_score_)
```

```python
from sklearn.ensemble import RandomForestClassifier

RFC = RandomForestClassifier()
params_RFC = {'n_estimators': range(700, 703)}
grid_RFC = GridSearchCV(RFC, params_RFC, cv=5, scoring='accuracy')
grid_RFC.fit(X_train, y_train)

print("Best n = %.4f" % grid_RFC.best_params_["n_estimators"])
print("Best accuracy = %.4f" % grid_RFC.best_score_)
```

```python
RFC = RandomForestClassifier(n_estimators=700)
params_RFC = {'criterion': ['gini', 'entropy'],
              'max_features': ['auto', 'sqrt', 'log2']}
grid_RFC = GridSearchCV(RFC, params_RFC, cv=5, scoring='accuracy')
grid_RFC.fit(X_train, y_train)

print("Best n =", grid_RFC.best_params_["criterion"])
print("Best n =", grid_RFC.best_params_["max_features"])
print("Best Accuracy = %.4f" % grid_RFC.best_score_)
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.model_selection import cross_validate

GBC = GradientBoostingClassifier()
GBC_err = []
err_min = 1

for i in range(100, 120):
    GBC.n_estimators = i
    cv_GBC = cross_validate(GBC,
                     X_train,
                     y_train,
                     scoring='accuracy',
                     cv=5
                    )
    GBC_err.append(1 - cv_GBC['test_score'].mean())
    if 1 - cv_GBC['test_score'].mean() < err_min:
        GBC_best = i
        err_min = 1 - cv_GBC['test_score'].mean()

print("Best n =", GBC_best)
print("Best Score =", 1-err_min)
```

```python
from catboost import CatBoostClassifier

ctb = CatBoostClassifier(early_stopping_rounds=30, iterations=181, learning_rate=0.15)
```

```python
def CV(X_train, y_train, number_folds, model, metrics):

    k = int(np.floor(float(X_train.shape[0]) / number_folds))

    errors = np.zeros(number_folds-1)
 
    for i in range(2, number_folds + 1):

        split = float(i-1)/i

        X = X_train[:(k*i)]
        y = y_train[:(k*i)]

        index = int(np.floor(X.shape[0] * split))
       
        X_trainFolds = X[:index]        
        y_trainFolds = y[:index]

        X_testFold = X[(index + 1):]
        y_testFold = y[(index + 1):]

        model.fit(X_trainFolds, y_trainFolds)
        errors[i-2] = metrics(model.predict(X_testFold), y_testFold)
    
    return errors.mean()
```

```python
CV(X_train, y_train, 10, ctb, accuracy_score)
```
